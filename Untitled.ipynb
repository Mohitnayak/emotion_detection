```python
import pandas as pd
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np
from scipy.sparse import hstack
import matplotlib.pyplot as plt

# === SET FEATURES ===
AUDIO_FEATURES = [col for col in train_df.columns if any(x in col for x in ['Pitch_', 'Loudness_', 'JitterShimmer_', 'HNR_', 'Formants_'])]
TEXT_FEATURE = 'Utterance'
TARGET = 'Emotion'

# === Prepare labels ===
y_train = train_df[TARGET]
y_dev = dev_df[TARGET]
y_test = test_df[TARGET]

# === Function to evaluate model ===
def evaluate_model(name, model, X_train, X_test, y_train, y_test, results_dict):
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    results_dict[name] = {
        "train_accuracy": accuracy_score(y_train, y_train_pred),
        "test_accuracy": accuracy_score(y_test, y_test_pred),
        "test_precision": precision_score(y_test, y_test_pred, average='weighted', zero_division=0),
        "test_recall": recall_score(y_test, y_test_pred, average='weighted', zero_division=0),
        "test_f1": f1_score(y_test, y_test_pred, average='weighted', zero_division=0),
    }

# === Store Results ===
results = {}

# === TEXT ONLY ===
tfidf = TfidfVectorizer(max_features=1000)
X_train_text = tfidf.fit_transform(train_df[TEXT_FEATURE])
X_test_text = tfidf.transform(test_df[TEXT_FEATURE])

evaluate_model("Text_RF", RandomForestClassifier(), X_train_text, X_test_text, y_train, y_test, results)
evaluate_model("Text_SVM", SVC(), X_train_text, X_test_text, y_train, y_test, results)

# === AUDIO ONLY ===
X_train_audio = train_df[AUDIO_FEATURES].fillna(0)
X_test_audio = test_df[AUDIO_FEATURES].fillna(0)

evaluate_model("Audio_RF", RandomForestClassifier(), X_train_audio, X_test_audio, y_train, y_test, results)
evaluate_model("Audio_SVM", make_pipeline(StandardScaler(), SVC()), X_train_audio, X_test_audio, y_train, y_test, results)

# === TEXT + AUDIO ===
X_train_combined = hstack([X_train_text, np.array(X_train_audio)])
X_test_combined = hstack([X_test_text, np.array(X_test_audio)])

evaluate_model("Text+Audio_RF", RandomForestClassifier(), X_train_combined, X_test_combined, y_train, y_test, results)
evaluate_model("Text+Audio_SVM", SVC(), X_train_combined, X_test_combined, y_train, y_test, results)

# === Convert results to DataFrame ===
results_df = pd.DataFrame(results).T.reset_index().rename(columns={"index": "Model"})

results_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split
import numpy as np


# Load train/dev/test sets
train_text_audio = pd.read_csv("pipe_data/train/meld_audio_prosody.csv")
dev_text_audio = pd.read_csv("pipe_data/dev/meld_audio_prosody.csv")
test_text_audio =  pd.read_csv("pipe_data/test/meld_audio_prosody.csv")

# Split into three types: text-only, audio-only, text+audio
def prepare_datasets(df, mode):
    X_text = df["Utterance"]
    X_audio = df.drop(columns=["Emotion", "Utterance", "utt_id"])
    y = df["Emotion"]

    if mode == "text":
        return X_text, y
    elif mode == "audio":
        return X_audio, y
    elif mode == "text+audio":
        return pd.concat([X_audio, X_text], axis=1), y
    else:
        raise ValueError("Invalid mode")

# Prepare training sizes
training_sizes = [100, 300, 500, 700, 1000, len(train_text_audio)]

# For each mode
results = {}
for mode in ["text", "audio", "text+audio"]:
    train_X_full, train_y_full = prepare_datasets(train_text_audio, mode)
    dev_X, dev_y = prepare_datasets(dev_text_audio, mode)

    rf_scores = {"train": [], "val": []}
    svm_scores = {"train": [], "val": []}

    for size in training_sizes:
        train_X = train_X_full[:size]
        train_y = train_y_full[:size]

        if mode == "text":
            rf_pipeline = Pipeline([
                ("tfidf", TfidfVectorizer()),
                ("clf", RandomForestClassifier())
            ])
            svm_pipeline = Pipeline([
                ("tfidf", TfidfVectorizer()),
                ("clf", SVC())
            ])
        elif mode == "audio":
            rf_pipeline = Pipeline([
                ("scale", StandardScaler()),
                ("clf", RandomForestClassifier())
            ])
            svm_pipeline = Pipeline([
                ("scale", StandardScaler()),
                ("clf", SVC())
            ])
        elif mode == "text+audio":
            tfidf = TfidfVectorizer()
            X_text_tfidf = tfidf.fit_transform(train_X["Utterance"])
            X_combined = np.hstack((train_X.drop(columns=["Utterance"]).values, X_text_tfidf.toarray()))

            dev_text_tfidf = tfidf.transform(dev_X["Utterance"])
            X_dev_combined = np.hstack((dev_X.drop(columns=["Utterance"]).values, dev_text_tfidf.toarray()))

            rf_pipeline = RandomForestClassifier()
            svm_pipeline = SVC()

            rf_pipeline.fit(X_combined, train_y)
            svm_pipeline.fit(X_combined, train_y)

            rf_train_acc = rf_pipeline.score(X_combined, train_y)
            svm_train_acc = svm_pipeline.score(X_combined, train_y)

            rf_val_acc = rf_pipeline.score(X_dev_combined, dev_y)
            svm_val_acc = svm_pipeline.score(X_dev_combined, dev_y)

            rf_scores["train"].append(rf_train_acc)
            rf_scores["val"].append(rf_val_acc)
            svm_scores["train"].append(svm_train_acc)
            svm_scores["val"].append(svm_val_acc)

            continue

        rf_pipeline.fit(train_X, train_y)
        svm_pipeline.fit(train_X, train_y)

        rf_train_acc = rf_pipeline.score(train_X, train_y)
        rf_val_acc = rf_pipeline.score(dev_X, dev_y)

        svm_train_acc = svm_pipeline.score(train_X, train_y)
        svm_val_acc = svm_pipeline.score(dev_X, dev_y)

        rf_scores["train"].append(rf_train_acc)
        rf_scores["val"].append(rf_val_acc)
        svm_scores["train"].append(svm_train_acc)
        svm_scores["val"].append(svm_val_acc)

    results[mode] = {"rf": rf_scores, "svm": svm_scores}

# Plotting
for mode in ["text", "audio", "text+audio"]:
    plt.figure(figsize=(10, 5))
    plt.plot(training_sizes, results[mode]["rf"]["train"], label="RF Train", marker='o')
    plt.plot(training_sizes, results[mode]["rf"]["val"], label="RF Val", marker='o')
    plt.plot(training_sizes, results[mode]["svm"]["train"], label="SVM Train", marker='x')
    plt.plot(training_sizes, results[mode]["svm"]["val"], label="SVM Val", marker='x')
    plt.title(f"Learning Curves ({mode})")
    plt.xlabel("Training Samples")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
